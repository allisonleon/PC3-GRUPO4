```python
#Paquetes que se usaran en el codigo
import requests
import json
from urllib.request import urlopen
from pandas import json_normalize
```

```python
url="https://www.balldontlie.io/api/v1/players" #url de la api que usaremos
uclient=urlopen(url) #abre el url
responsehtml=uclient.read() #lee el url almacenado en uclient
uclient.close() # cierra el url
data = responsehtml.decode() # decodifica el codigo
#type(data) # si le aplicamo un type a data nos dara de resultado 'str'
data
```

```python
info = json.loads(data)
print(type(info))
info
#ORDENAR LA INFORMACIÓN,CONVERTIR LA DATA A FORMATO JSON.
```

```python
#type(info['data']) ##list
len(info['data']) ## 25 elementoss de la data
basq=info['data'] # Se almacena a la variable basq
from pandas import json_normalize
df = json_normalize(basq) #el json lo vuelve un dataframe
df2=df.drop(['id','height_feet','height_inches','weight_pounds'], axis=1)
df2
```

```python
#renombrado de columnas
#cabecera=['Nombre','Apellido','Posicion','ID_Equipo',"Abreviado_del_equipo",'Ciudad','Zona_equipo','Division','Nombre_Oficial_Equipo','Nombre_Equipo']
df2.columns=['Nombre','Apellido','Posicion','ID_Equipo',"Abreviatura",'Ciudad','Asociacion','Division','Nombre_Oficial_Equipo','Nombre_Equipo']
df2
```

```python
#GUARDAMOS EL DATA FRAME EN UN ARCHIVO .CSV
df2.to_csv('Api_basquet.csv', index = False, header = True, encoding ="utf-8")
```
