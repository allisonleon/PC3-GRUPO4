```python
url="https://www.balldontlie.io/api/v1/players" #url de la api que usaremos
uclient=urlopen(url) #abre el url
responsehtml=uclient.read() #lee el url almacenado en uclient
uclient.close() # cierra el url
data = responsehtml.decode() # decodifica el codigo
#type(data) # si le aplicamo un type a data nos dara de resultado 'str'
data
```
